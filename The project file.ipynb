# Importing all the necessary Libraries

```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt


```

# A list of all the Mean Squared Errors

```python
MSEList = []
```

# Importing the data already downloaded

```python
data = pd.read_csv("concrete_data.csv")
```

# Splitting the data into Features and Labels

```python
Labels = data['Strength'].values
Features = data.drop('Strength', axis=1).values
```

# Splitting the data into Train and Test data

```python
Features_train, Features_test, Labels_train, Labels_test = train_test_split(Features, Labels, test_size=0.3, shuffle=True)
```

# Making the regression Model

```python
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(len(Features[0]),)))
model.add(Dense(1))

model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(Features_train, Labels_train, epochs=50)
```

# Predicting the data 

```python
Predictions = model.predict(Features_test)

MeanSqError = mean_squared_error(Labels_test, Predictions)

print(MeanSqError)
```

# Making a loop to save all the MSE Values and Running the Loop for 50 Times

```python
for i in range(50):
    Features_train, Features_test, Labels_train, Labels_test = train_test_split(Features, Labels, test_size=0.3, shuffle=True)

    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(len(Features[0]),)))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mse')
    model.fit(Features_train, Labels_train, epochs=50)

    Predictions = model.predict(Features_test)
    MeanSqError = mean_squared_error(Labels_test, Predictions)

    print(MeanSqError)
    MSEList.append(MeanSqError)

```

# Calculating Mean and Standard Deviation of All the Mean Squred Error Values

```python
mean = np.mean(MSEList)
standard_deviation = np.std(MSEList)

print(f'''Mean: {mean}
Standard Deviation: {standard_deviation}''')
```
